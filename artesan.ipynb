```python
# imports
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('VendasSet2020.csv')
df1
```

```python
df2 = pd.read_csv('VendasNov2020.csv')
df2
```

```python
df1.rename(columns={'Descrição do Produto (completa)':'descricao_produto',
                    'Quantidade':'qtd',
                    'Total de Mercadoria':'valor_mercadoria',
                    'Frete':'frete',
                    'Total da Nota Fiscal':'valor_nota_fiscal'}, inplace = True)


df2.rename(columns={'descricao_do_produto_completa':'descricao_produto',
                    'quantidade':'qtd',
                    'total_de_mercadoria':'valor_mercadoria',
                    'total_da_nota_fiscal':'valor_nota_fiscal'}, inplace = True)
```

```python
df1.corr()
```

```python
plt.scatter(df1['valor_mercadoria'],df1['qtd'])
plt.scatter(df2['valor_mercadoria'],df2['qtd'])
plt.show()
```

```python
sns.pairplot(df1, vars=["valor_mercadoria","valor_nota_fiscal"], hue='qtd')
sns.pairplot(df2, vars=["valor_mercadoria","valor_nota_fiscal"], hue='qtd') 
```

```python
sns.heatmap(df1.corr(), cbar=True)
plt.show
```

```python
sns.heatmap(df2.corr(), cbar=True)
plt.show
```
